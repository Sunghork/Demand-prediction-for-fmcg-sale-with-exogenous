```python
# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Summary statistics of the dataset
summary_statistics = df.describe()
print(summary_statistics)

# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

# Check data types
data_types = df.dtypes
print(data_types)

# Plot sales distribution
plt.figure(figsize=(12, 4))
sns.histplot(df['sales'], kde=True, bins=30)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

# Plot sales over time
plt.figure(figsize=(17, 4))
plt.plot(df['date'], df['sales'], label='Sales')
plt.title('Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Plot sales on promotion vs. non-promotion days
plt.figure(figsize=(12, 4))
sns.boxplot(x='promotion', y='sales', data=df)
plt.title('Sales on Promotion vs. Non-Promotion Days')
plt.xlabel('Promotion')
plt.ylabel('Sales')
plt.xticks([0, 1], ['Non-Promotion', 'Promotion'])
plt.show()

# Plot sales on holiday vs. non-holiday days
plt.figure(figsize=(12, 4))
sns.boxplot(x='holiday', y='sales', data=df)
plt.title('Sales on Holiday vs. Non-Holiday Days')
plt.xlabel('Holiday')
plt.ylabel('Sales')
plt.xticks([0, 1], ['Non-Holiday', 'Holiday'])
plt.show()

# Calculate the Spearman correlation matrix
spearman_corr = df.corr(method='spearman')

# Plot the Spearman correlation matrix
plt.figure(figsize=(25, 12))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation Matrix')
plt.show()

# Perform Augmented Dickey-Fuller test
from statsmodels.tsa.stattools import adfuller

# Perform the ADF test on sales data
adf_result = adfuller(df['sales'])

# Extract and format the results
adf_statistic, p_value, used_lag, n_obs, critical_values, icbest = adf_result
adf_summary = {
    'ADF Statistic': adf_statistic,
    'p-value': p_value,
    'Used Lag': used_lag,
    'Number of Observations': n_obs,
    'Critical Values': critical_values,
    'IC Best': icbest
}
print(adf_summary)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller

# Load the dataset
df = pd.read_csv('D:\Python\demand-prediction\Dataset\SKU4.csv')

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Extract day of week, month, and year features
df['day_of_week'] = df['date'].dt.dayofweek  # Monday=0, Sunday=6
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year

# Split the dataset into train, validation, and test sets
train_size = int(len(df) * 0.7)
val_size = int(len(df) * 0.15)

train = df.iloc[:train_size]
val = df.iloc[train_size:train_size + val_size]
test = df.iloc[train_size + val_size:]

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Summary statistics of the dataset
summary_statistics = df.describe()
print(summary_statistics)

# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

# Check data types
data_types = df.dtypes
print(data_types)

# Plot sales distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['sales'], kde=True, bins=30)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

# Plot sales over time
plt.figure(figsize=(20, 4))
plt.plot(df['date'], df['sales'], label='Sales')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Calculate the Spearman correlation matrix
spearman_corr = df.corr(method='spearman')

# Plot the Spearman correlation matrix
plt.figure(figsize=(27, 10))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation Matrix with Day of Week, Month, and Year')
plt.show()

# Perform Augmented Dickey-Fuller test on sales data
adf_result = adfuller(df['sales'])

# Extract and format the results
adf_statistic, p_value, used_lag, n_obs, critical_values, icbest = adf_result
adf_summary = {
    'ADF Statistic': adf_statistic,
    'p-value': p_value,
    'Used Lag': used_lag,
    'Number of Observations': n_obs,
    'Critical Values': critical_values,
    'IC Best': icbest
}
print(adf_summary)
```
